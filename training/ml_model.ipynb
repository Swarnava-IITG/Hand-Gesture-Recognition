```python
import tensorflow as tf
import numpy as np
import keras
import pandas as pd
```

```python
data_file=pd.read_csv("data.csv")
data_file = data_file.sample(frac=1).reset_index(drop=True)
total_recognitions=5
```

```python
data = data_file.drop("label",axis=1).to_numpy()
label = data_file["label"].to_numpy()
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(40,1)),
    keras.layers.Dense(128,activation="relu"),
    keras.layers.Dense(total_recognitions,activation="softmax")
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(data,label,epochs=5)
```

```python
tdata_file=pd.read_csv("tdata.csv")
tdata_file = tdata_file.sample(frac=1).reset_index(drop=True)

tdata = tdata_file.drop("label",axis=1).to_numpy()
tlabel = tdata_file["label"].to_numpy()
```

```python
tloss,tacc = model.evaluate(tdata,tlabel)
print(tloss,tacc)
```

```python
model.save("gesturedetector.keras")
```
